`````python
# Plot val_dice for the most recent run
import glob, pandas as pd, matplotlib.pyplot as plt
files = sorted(glob.glob('results/NN_UNet2D_MedSeg_*_publication.csv'))
if files:
    last = files[-1]
    df = pd.read_csv(last)
    if 'val_dice' in df.columns:
        plt.figure(figsize=(6,4))
        plt.plot(df['epoch'], df['val_dice'], marker='o')
        plt.title(f'Validation Dice per Epoch\n{last}')
        plt.xlabel('epoch'); plt.ylabel('val_dice'); plt.grid(True, alpha=0.3)
        plt.show()
    else:
        print('No val_dice column in last results file.')
else:
    print('No results files found yet.')
`````
